## Basic Example

To get a feel for the language, let us conclude with a basic example of using the *SciPy Stack*. We will work with a simple Ordinary Least Squares (OLS) model.

$$Y = X \beta + \epsilon$$

We start by simulating some sythetic dataset. Then we visualize the relationship and fit a basic OLS regression.

But first, some housekeeping. For additional information, check out the [cell magics](http://nbviewer.ipython.org/github/ipython/ipython/blob/1.x/examples/notebooks/Cell%20Magics.ipynb) example notebook.

```python
%%capture
# Capture the stdout/err of any block of Python code

%pylab inline
# Provide plots inline
```

### Simulate Synthetic Sample

```python
# Import relevant libraries from the SciPy Stack
import numpy as np

# Specify parametrization
num_agents = 1000
num_covars = 3

betas_true = np.array([0.22, 0.30, -0.1]).T

# Set a random seed to ensure repcomputability in light of randomness
np.random.seed(123)

# Sample exogenous agent characteristics from a uniform distribution
X = np.random.rand(num_agents, num_covars)

# Sample random disturbances from a standard normal distribution and rescale
eps = np.random.normal(scale=0.1, size=num_agents)

# Construct endogenous agent characteristic
Y = 0.1 + np.dot(X, betas_true) + eps
```

### Visualize Relationship

```python
# Import relevant libraries from the SciPy Stack
import matplotlib.pyplot as plt

# Initialize canvas
ax = plt.figure(figsize=(12,8)).add_subplot(111, axisbg='white')

# Plot synthetic sample
ax.plot(np.dot(X, betas_true), Y, 'o')

# Set axis labels and ranges
ax.set_xlabel(r'$X\beta$', fontsize=20)
ax.set_ylabel(r'$Y$', fontsize=20)

# Remove first element on y-axis
ax.yaxis.get_major_ticks()[0].set_visible(False)

# Add title
plt.suptitle('Synthetic Sample', fontsize=20)

# Show plot
plt.show()
```

### Fit OLS Model

```python
# Import relevant libraries from the SciPy Stack
import statsmodels.api as sm

# Add an intercept
X = sm.add_constant(X)

# Specify and fit the model
rslt = sm.OLS(Y, X).fit()

# Provide some summary
print(rslt.summary())
```
